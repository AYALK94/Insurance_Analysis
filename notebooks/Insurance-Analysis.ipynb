```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
df = pd.read_csv('/home/ayalk94/Documents/GitHub/Insurance_Analysis/data/AB Hypothesis Testing/data/insurance_data.csv')

# Overview of the dataset
print(df.head())
print(df.info())
```

### Check for Missing Data:

```python
# Check for missing values
print(df.isnull().sum())
```

```python
df.columns
```

### Univariate Analysis

```python

# Assuming df['Claimed'] is a binary column (e.g., 0 for not claimed, 1 for claimed)

# 1. Bar plot for Province
df['Province'].value_counts().plot(kind='bar', figsize=(8, 5))
plt.title("Distribution of Claims by Province")
plt.xlabel("Province")
plt.ylabel("Count")
plt.show()

# 2. Bar plot for Gender
df['Gender'].value_counts().plot(kind='bar', figsize=(6, 4))
plt.title("Distribution of Claims by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# 3. Pie chart for Claim Status (Claimed or Not Claimed)
df['Claimed'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Claimed vs Not Claimed")
plt.ylabel("")  # Hide the y-label for better aesthetics
plt.show()
```

### Descriptive Statistics for the Claimed Column

```python
# Descriptive statistics for Claimed column (if it's numerical or binary)
print(df['Claimed'].describe())
```

### Descriptive Statistics for Categorical Data

```python
# Summary for Province
print(df['Province'].value_counts())

# Summary for Gender
print(df['Gender'].value_counts())
```

### Bivariate and Multivariate Analysis

### Scatter Plots

```python
# Bar plot to show claims by Gender
sns.barplot(x='Gender', y='Claimed', data=df, ci=None)
plt.title("Claims by Gender")
plt.ylabel("Average Claimed (0 or 1)")
plt.show()
```

### Bar Plot for Claims by Province:

```python
sns.barplot(x='Province', y='Claimed', data=df, ci=None)
plt.title("Claims by Province")
plt.ylabel("Average Claimed (0 or 1)")
plt.xticks(rotation=45)
plt.show()
```

### Outlier Detection

```python
sns.countplot(x='Claimed', data=df)
plt.title("Count of Claimed vs Not Claimed")
plt.show()
```

### Visualization

### Distribution of Claims by Gender

```python
sns.countplot(x='Gender', hue='Claimed', data=df)
plt.title("Distribution of Claims by Gender")
plt.show()
```

### Distribution of Claims by Province

```python
sns.countplot(x='Province', hue='Claimed', data=df)
plt.title("Distribution of Claims by Province")
plt.xticks(rotation=45)
plt.show()
```

### Pie Chart for Claim Proportions:

```python
df['Claimed'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Proportion of Claims vs Not Claimed")
plt.ylabel('')  # Remove ylabel for aesthetics
plt.show()
```
